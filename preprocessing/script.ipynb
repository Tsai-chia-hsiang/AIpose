```python
import os
import torch
import numpy as np
from tqdm import tqdm
import json
from sentence_transformers import SentenceTransformer, util
```

```python
labels=[]
with open("format_label.json", "r") as f:
    labels = json.load(f)
labels
```

```python
label_des = [i[1] for i in labels]
label_des
```

```python
ptype= []
with open("prototype.txt", "r") as f:
    for i in f.readlines():
        li = i.strip()
        ptype.append(li)
ptype
```

```python
model = SentenceTransformer(
    os.path.join("sentence-transformers_all-mpnet-base-v2"),
    device='cpu'
)
```

```python
label_emd =model.encode(label_des)
```

```python
ptype_ebd = model.encode(ptype)
```

```python
cossim = util.cos_sim(label_emd, ptype_ebd)
print(cossim.size())
```

```python
cluster_id = cossim.argmax(dim=1)
cluster_id = cluster_id.tolist()
```

```python
clustering = {k:[] for k in ptype}

for i, ci in enumerate(cluster_id):
    clustering[ptype[ci]].append(labels[i]) 

with open("clustering.json", "w+") as f:
    json.dump(clustering, f, indent=4, ensure_ascii=False)
```

```python
for k in clustering:
    print(k, end=" ")
    print(len(clustering[k]))
```
